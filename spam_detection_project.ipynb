```python
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk import TweetTokenizer
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
sw = stopwords.words('english')
tk = TweetTokenizer()
le=WordNetLemmatizer()
def preprocess_text(text):
    if not isinstance(text, str):
        return ""
    text = re.sub("[^0-9a-zA-Z]+", " ", text)
    tokens = tk.tokenize(text)
    tokens = [le.lemmatize(i.lower()) for i in tokens]
    tokens = [i for i in tokens if i not in sw]
    return " ".join(tokens)
tweets_clean = df['v2'].apply(preprocess_text)
print("First 10 cleaned tweets:\n")
print(tweets_clean.head(10))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
y=tf.fit_transform(tweets_clean)
print(y)
print(tf.get_feature_names_out())
print(y.toarray())
```

```python
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
clf=Pipeline([("tfidf",TfidfVectorizer()),("naive bayes",MultinomialNB())])
clf.fit(tweets_clean, df.iloc[:,0])
```
